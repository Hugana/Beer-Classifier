## Introdução Teórica

### Pré-Processamento de Dados

Quando se lida com dados reais e habitualmente necessário proceder a um passo de pre-processamento dos mesmos antes se de aplicar técnicas de aprendizagem automatica. Dados como texto, imagem, audio, e muitos outros, apresentam por vezes valores omissos, pontos espurios, variáveis com escalas e valores medios distintos.
<br>
São exemplos de processos de pre-processamento de dados:
- Sampling - dividir os dados de um grupo grande
- Normalização e escalar dados - por vezes os dados podem estar em diferentes dimensões, pelo que é necessário restringi-los a valores mais próximos uns dos outros
- Extração de caracteristicas - métodos como PCA (Análise de componentes principais) e LDA (Análise de discriminantes principais) ajudam a reduzir a dimensionalidade dos dados mantendo uma boa eficácia. nas classificações.
<br><br>

Mais específicamente, com dados de texto:
- Stemming - é uma abordagem que visa reduzir palavras para as formas "radicais", removendo sufixos para representar a forma raiz da palavra.
- Limpeza - remover elementos indesejados ou irrelevantes do texto, tal como números e caractéres especiais, simplificando assim a análise e melhorando a eficácia dos classificadores.
- Vetorização - conversão de dados em formato de texto, como palavras ou documentos, em vetores numéricos que podem ser compreendidos por algoritmos de aprendizado de máquina.

### Métodos supervisionados vs não supervisionados

A __aprendizagem supervisionada__ trata-se de um aprendizagem em que o classificador é treinado com um conjunto de dados reais(desejados), ou seja, forcece-se já a resposta certa juntamente com o conjunto de dados de treino. Quando é fornecido um novo conjunto de dados, através do que o classificador já aprendeu, consegue comparar e obter um conjunto de dados estimados, que depois podem ser comparados aos desejados.
<br>
A aprendizagem supervisionada pode ser dividida principalmente em duas categorias: classificação e regressão. Essa divisão é baseada no tipo de saída que o método tenta prever.
- __Classificação__ - o método é treinado para prever a qual classe ou categoria uma entrada pertence. As classes são pré-definidas e discretas.
- __Regressão__ - o modelo é treinado para prever um valor contínuo. Em vez de prever classes, o objetivo é estimar um valor numérico.
<br>

Exemplos destes métodos:
- __Regressão__
- __Regressão Logística__
- __Classificadores por distância__
<br><br>

Já a __aprendizagem não supervisionada__, ao contrário da aprendizagem supervisionada, lida com dados que não são conhecidas as classes ou categorias. Os métodos são deixadso para encontrar padrões ou estrutura nos dados por conta própria.<br>
Exmploes de métodos que utilizam esta aprendizagem são, por exemplo, o __PCA__ e o __Clustering__.


### Métricas de Distância

Do ponto de vista matemático, distância é um conceito mais geral e abstrato, quando comparamos com a distância percessionada por humanos, que abrange não só a distância euclideana, bem como um grande número de outras métricas. <br>
Métricas de distância são uma ferramenta essencial para diversos tópicos de aprendizagem automática, entre os quais técnicas de regressão, classificação, e métodos de agrupamentos.
<br>
Várias técnicas de classificaçãao são direta ou indiretamente baseadas em medidas de distância. Dois dos métodos de classificação mais simples são:
- __Classificador de distancia ao centroide__
- __Classificador do k vizinhos mais proximos (k-NN)__

<br>

Na métrica de distância ao centróide é calculada usando a distância euclidiana entre um ponto e o centróide, neste caso da categoria. <br>

No contexto de k-NN, a métrica de distância é fundamental para determinar os "vizinhos mais próximos". A ideia é calcular a distância entre um ponto de consulta e todos os outros pontos no conjunto de dados, identificar os k pontos mais próximos e tomar uma decisão com base nas classes ou valores associados a esses vizinhos mais próximos.

#### Limpeza de dados

Começa-se por criar uma função auxiliar para aplicar a limpeza do texto. Por exemplos visualizados anteriormente, não foram encontrados caracteres especiais, parece não ser necessário fazer conversão para, por exemplo, UTF-8. Por outro lado, faz sentido remover números e outro caracteres como o underscore, que não terão qualquer interesse para fazer a classificação.

```python
def clean_text(text):
    
    # remover numeros e caracters especiais
    cleaned_text = re.sub(r'[^a-zA-Z\u00C0-\u00FF]',' ', text)
    # coverter para lower case
    cleaned_text = cleaned_text.lower()
    
    return cleaned_text
```

#### Stemming

Esta técnica é o processo de transformar uma palavra na sua raiz, o que permite mapear palavras semelhantes numa única palavra, como por exemplo, studies, studying, studied ⇒ studi.<br>
Neste trabalho, usar-se-á 3 tipos de stemming:
- __Lancaster__
- __Snowball__
- __Porter__
<br>

O __Porter__ é conhecido por ser relativamente simples e rápido, mas pode ser agressivo em algumas situações, resultando em imprecisões. É baseado em uma série de regras heurísticas que são aplicadas de maneira iterativa para reduzir as palavras à sua forma raiz.<br>
O __Lancaster__ utiliza um conjunto diferente de regras e é mais eficiente em termos de redução do comprimento das palavras. Por ser mais agressivo, o Lancaster pode gerar raízes mais curtas, mas pode também ser mais impreciso em algumas circunstâncias.
<br>
O __Snowball__ foi também desenvolvido Porter, o mesmo criador do Porter Stemmer, como uma melhoria deste. É mais modular e flexível do que o original, permitindo ajustes específicos para diferentes contextos linguísticos.

Começa-se por criar uma função auxiliar para aplicar o stemming aos Docs (texto das reviews).

```python
# funcao para aplicar stemming
def apply_stemming(cleanedDocs, stemmer):
    
    stemmedDocs = []
    
    for doc in cleanedDocs:
        
        # separar as palavras
        words = doc.split()
        
        # lista vazia para adicionar
        stemmed_words = []
        
        for word in words:
            # adicionar as palavras numa lista
            stemmed_words.append(stemmer.stem(word))
        
        # juntar de novo as palavras para formar o doc
        stemmed_review = ' '.join(stemmed_words)
        
        # manter o original
        stemmedDocs.append(stemmed_review)

    return stemmedDocs
```

E aplica-se os 3 tipos de Steeming para comparar os seus resultados

```python
#LancasterStemmer
Doc_lancaster_stem = apply_stemming(Docs_treino_cleaned, LancasterStemmer())

#SnowballStemmer
Doc_snowball_stem = apply_stemming(Docs_treino_cleaned, SnowballStemmer(language="english"))

#PorterStemmer
Doc_porter_stem = apply_stemming(Docs_treino_cleaned, PorterStemmer())
```

#### Vetorização do vocabulário 

Consiste na converção de documentos de texto em representações numéricas, chamadas vetores, que podem ser facilmente manipuladas por algorítmos de classificação.<br>
Neste trabalho, faz-se uso do método __td-idf__ representa em forma a matriz de contagens documento-termo (document-token), dando mais relevância a termos mais informativos. A intuição por detrás do méetodo é dar um peso maior a termos que aparecem frequentemente num dado documento, mas não em muitos documentos do corpus. Se uma palavra aparece em poucos documentos, mas muitas vezes num em particular, entao o pressuposto é que essa palavra descreve bem o contéudo desse documento.<br>
Por padrão, o output deste método já é normalizado, fazendo assim o passo do escalamento e normalização "por nós".

```python
def getTfidfVectorizer(Docs, min_df_array, token_pattern_array):
    
    len_tokens = []
    
    min_df_data = []
    token_pattern_data = []
    
    
    for i in range(len(min_df_array)):
        for j in range(len(token_pattern_array)):
            tfidf = TfidfVectorizer(min_df=min_df_array[i], token_pattern=token_pattern_array[j]).fit(Docs)
            len_tokens.append(len(tfidf.get_feature_names()))
            min_df_data.append(i + 1)
            token_pattern_data.append(j + 1)
            
    return len_tokens, min_df_data, token_pattern_data
```

```python
min_df_array = [1, 2, 3, 4, 5]
token_pattern_array = [r'\b\w+\b', r'\b\w\w+\b', r'\b\w\w\w+\b']



len_tokens_Lancaster, min_df_data, token_pattern_data  = getTfidfVectorizer(Doc_lancaster_stem,min_df_array, token_pattern_array)

len_tokens_Snowball, _, _  = getTfidfVectorizer(Doc_snowball_stem,min_df_array, token_pattern_array)

len_tokens_Porter, _, _    = getTfidfVectorizer(Doc_porter_stem,min_df_array, token_pattern_array)
```

É também importante preparar os dados para a classificação binária. <br>
Nesta classificação, uma cerveja e considerada muito boa quando obteve uma pontuação global de 9 ou mais valores e considerada muito má quando obteve uma pontuação global de 2 ou menos valores.<br> 
Há duas prespetivas possiveis de testar nesta classificação:
- Retira-se todos os dados com classificações que não seja abrangidas pelas condições acima, e testa-se apenas se uma review é ou muito boa ou muito má.
- Faz-se duas classificações, uma para classificar todos os dados como muito más ou não muito más, e outra para classificar os dados como muito boas ou não muito boas.

A primeira opção não parece viável em vários sentidos, visto que com 4 classifcações, os dados obtidos parecem ser reduzidos para fazer uma boa classifcação, e também porque faz mais sentido conseguir aplicar este classificador a todos os dados e conseguir perceber se uma review foi muito boa, muito má ou nenhum dos dois. <br>
O grupo escolheu então a segunda opção.

Faz-se então dois grupos de "target", uma para classificar as reviews como muito más e não muito más, onde:
- 1 serão os overalls 1 e 2
- 0 serão os restantes

Para classifcar as reviews como muito boas e não muito boas:
- 1 serão os overalls com 9 e 10
- 0 serão os restantes


```python
target_treino_negativo = [0 if num >= 3 else 1 for num in target_treino]    
target_teste_negativo = [0 if num >= 3 else 1 for num in target_teste]

target_treino_positivo = [0 if num <= 8 else 1 for num in target_treino]    
target_teste_positivo = [0 if num <= 8 else 1 for num in target_teste]
```

Ainda aqui, vamos introduzir mais um parametro na vectorização, que é o n_grams. O n_grams é uma maneira de capturar (parte) da informação contextual e incluir na representação BoW sequências de duas ou mais palavras que habitualmente apareçam nos documentos. Conjuntos de duas palavras são denominados bi-gramas e de três palavras, tri-gramas.

Para auxiliar, cria-se uma função para obter os scores com os vários valores de min_df, token_pattern e n_grams. Há vários métricas que permitem avaliar esta classificação binária, entre eles: <br>
- Cross Validation
- Score (precisão)
- Curvas ROC
- Curvas precision/recall
- Menor número de erros <br>

Sendo assim, a melhor forma de avaliar é fazer uma média do melhor resultado de cada avaliação.



```python
def getScoresTfidfVectorizer(Docs_train, Docs_test, target_train, target_test, min_df, token_pattern, n_grams):
    
    #Separar os docs de treino, em treino e validação, preparando assim estes dados para fazer o ajuste dos hiperparametros
    X_train, X_valid, y_train, y_valid = train_test_split(Docs_train, target_train, test_size=1/3, stratify=target_train)
    
    tfidf=TfidfVectorizer(min_df=min_df, token_pattern=token_pattern, ngram_range=n_grams).fit(X_train) 
    
    X_treino = tfidf.transform(X_train)
    X_valid = tfidf.transform(X_valid)
    X_teste = tfidf.transform(Docs_test)
    
    
    CArray = [0.01, 0.1, 1, 10]
    penaltyArray = ['l1', 'l2']
    
    best_cv_score = 0
    best_roc_auc = 0
    best_pr_auc = 0
    best_menor_erros = 75000
    
    best_par_cv_score = best_par_roc_auc = best_par_pr_auc = best_par_menor_erros = [0,'']
    
    for C in CArray:
        for pen in penaltyArray:
            lr = LogisticRegression(penalty=pen, solver='liblinear', C=C, max_iter=5000, tol=1e-3)
            lr.fit(X_treino, y_train)
            
            ye_valid = lr.predict(X_valid)

            # Cross Validation
            cv_score_mean = np.mean(cross_val_score(lr, X_treino, y_train, cv=5))

            # Curvas ROC
            fpr, tpr, thresholds = roc_curve(y_valid, ye_valid)
            roc_auc = auc(fpr, tpr)

            # Curvas precision/recall
            precision, recall, thresholds_pr = precision_recall_curve(y_valid, ye_valid)
            pr_auc = auc(recall, precision)

            # Menor número de erros
            menor_num_erros = np.sum(y_valid != ye_valid)
            
            if cv_score_mean > best_cv_score:
                best_cv_score = cv_score_mean
                best_par_cv_score = [pen,C]
                
            if roc_auc > best_roc_auc:
                
                best_roc_auc = roc_auc
                best_par_roc_auc = [pen,C]
                
            if pr_auc > best_pr_auc:
                best_pr_auc = pr_auc
                best_par_pr_auc = [pen,C]
                
            if menor_num_erros < best_menor_erros:
                best_menor_erros = menor_num_erros
                best_par_menor_erros = [pen,C]
         
    bestParamns = [best_par_cv_score,
                  best_par_roc_auc,
                  best_par_pr_auc,
                  best_par_menor_erros]
    
    print('CVScore', best_par_cv_score, 'ROC', best_par_roc_auc, 'PR', best_par_pr_auc, 'Menor Erros', best_par_menor_erros)
    
    scores = []
    
    for param in bestParamns:
        lr = LogisticRegression(penalty=param[0], solver='liblinear',C=param[1], max_iter=5000, tol=1e-3)
        lr.fit(X_treino, y_train)
        
        scoreTrain = lr.score(X_treino, y_train)
        scoreVal   = lr.score(X_valid, y_valid) 
        scoreTest  = lr.score(X_teste, target_test)
        
        scores.append(scoreTrain)
        scores.append(scoreVal)
        scores.append(scoreTest)
        
    
    print('Score:',np.mean(scores))
    
    
    return np.mean(scores), len(tfidf.get_feature_names())
                
```

```python
ngrams_array      = [(1,1), (1,2), (1,3)]
min_df_array = [1, 2, 3, 4, 5]
token_pattern_array = [r'\b\w+\b', r'\b\w\w+\b', r'\b\w\w\w+\b']

resultados_criticas_muito_mas={}
count = 0

for ngrams in ngrams_array:
    for min_df in min_df_array:
        for token_pattern in token_pattern_array:
            
            scores_len = getScoresTfidfVectorizer(Docs_train, Docs_test,target_treino_negativo, target_teste_negativo, min_df, token_pattern, ngrams)
            
            aux = {'n_grams' : ngrams,
                  'min_df' : min_df,
                  'token_pattern' : token_pattern,
                  'scores' : scores_len[0],
                  'len' : scores_len[1]}
            
            resultados_criticas_muito_mas[str(count)] =  aux
            
            count += 1
            
```

```python
def getBestParamsLogistic(ficheiro, ficheiro_saida):
    
    D = pickle.load(open(ficheiro,'rb'))

    X_treino = D['X_treino']
    target_treino_muito_mas = D['target_treino']
    
    CArray = [0.01, 0.1, 0.2 ,0.5, 1, 2.5, 5]
    penaltyArray = ['l1', 'l2']

    best_score = 0
    best_C = best_penalty = None
    
    dicionarios={}
    count = 0
    for C in CArray:
        for pen in penaltyArray:
            lr = LogisticRegression(penalty=pen, solver='liblinear', C=C, max_iter=5000, tol=1e-3)
            lr.fit(X_treino, target_treino_muito_mas)

            
            score = np.mean(cross_val_score(lr, X_treino, target_treino_muito_mas, cv=5))
                
            print("C: " + str(C) + ",Penalty: " + str(pen) + ",Score:" + str(score))
            resultados={'C' : C,
               'pen' : pen,
               'score': score}
            
            dicionarios[str(count)] = resultados
            count += 1
                
                
    pickle.dump(dicionarios, open(ficheiro_saida, 'wb'))
    
            
        
```

```python

lr = LogisticRegression(penalty='l2', solver='liblinear', C=2.5)

D = pickle.load(open('dados_treino_teste_muito_boas.p','rb'))

X_treino = D['X_treino']
y_treino = D['target_treino']

X_treino, X_valid, y_treino, y_valid = train_test_split(X_treino, y_treino, test_size=1/3, stratify=y_treino)

lr.fit(X_treino, y_treino)
            
ye_valid = lr.predict(X_valid)

# Curvas ROC
fpr, tpr, thresholds = roc_curve(y_valid, lr.decision_function(X_valid))
roc_auc = auc(fpr, tpr)

# Menor número de erros
menor_num_erros = np.sum(y_valid != ye_valid)

matriz = confusion_matrix(y_valid,ye_valid)
```

```python
# Pontos operacionais:
# TP-Rate -> TP/(TP+FN) e FP-Rate -> FP/(FP+TN)
p1 = matriz[0,1]/(matriz[0,1] + matriz[0,0])
p2 = matriz[1,1]/(matriz[1,1] + matriz[1,0])


plt.figure(figsize=(7,7))
# Plot da Curva ROC
plt.plot(fpr,tpr, color='darkblue', label='Curva ROC')

# Plot dos Pontos
plt.plot(p1, p2, 'o', color='darkorange',     label='Ponto Operacional')

plt.title('Curva ROC e Pontos Operacionais')
plt.ylabel('Taxa de True Positive')
plt.xlabel('Taxa de False Positive')
plt.legend()
plt.axis('scaled')
plt.show()

print("Matriz de confusão:")
print(matriz)
print("Taxa de erros:")
print(menor_num_erros/len(y_valid))
```

__Como é observável na curva ROC__, o ponto operacional está junto ao canto inferior esquerdo, e como sabemos o ponto ótimo seria no canto superior esquerdo,ou seja, quer-se que a taxa de verdadeiros positivos seja o maior possível, e a taxa de falsos positivos a menor possível.<br>
Também observando em conjunto a matriz de confusão, é possivel perceber que a taxa de falsos positivos é reduzida, mas leva a uma taxa também reduzida de verdadeiros positivos. <br>
Para maximizar a taxa de verdadeiros positivos, pode-se __calibrar__ este classificador, mas trará algumas consequências, ,tal com:
- __aumeto número de falsos positivos__
- __aumento dos erros__

```python
y2c = lr.decision_function(X_valid)
#print(y2c)
minLimear = np.min(y2c)
maxLimear = np.max(y2c)

print('Valores de decisão:')
print('Min:', minLimear, 'Max:', maxLimear)

dist_menor = 1000
v = 0
for l in (np.arange(minLimear, maxLimear, 0.001)): 
    mat = confusion_matrix(y_valid,y2c >= l)
    p1 = mat[0,1]/(mat[0,1] + mat[0,0])
    p2 = mat[1,1]/(mat[1,1] + mat[1,0])
    distancia = math.dist([p1,p2], [0,1])
    if (distancia<dist_menor):
        dist_menor = distancia
        v = l                                                                    
```

```python
matrizC = confusion_matrix(y_valid, y2c >= v)

# Obtenção dos pontos operacionais:
# TP-Rate -> TP/(TP+FN) e FP-Rate -> FP/(FP+TN)
p1 = matrizC[0,1]/(matrizC[0,1] + matrizC[0,0])
p2 = matrizC[1,1]/(matrizC[1,1] + matrizC[1,0])


plt.figure(figsize=(7,7))
# Plot da Curva ROC
plt.plot(fpr,tpr, color='darkblue', label='Curva ROC')

# Plot dos Pontos
plt.plot(p1, p2, 'o', color='darkorange',     label='Ponto Operacional')

plt.title('Curva ROC e Pontos Operacionais')
plt.ylabel('Taxa de True Positive')
plt.xlabel('Taxa de False Positive')
plt.legend()
plt.axis('scaled')
plt.show()

# Mostrar resultados
print('Nº total de erros: ', np.sum(y_valid != (y2c >= v))/len(y_valid))
print("Limiar de decisão:", v)
print('Nova matriz de confusão:')
print(matrizC)
```

## SVM (Support Vector Machine)

Maquinas de suporte vetorial (Support Vector Machine), são algoritmos supervisionados de aprendizagem automatica aplicáveis a problemas de classificação e regressão. Foi inicialmente desenvolvido para problemas de classificação binária.

Os objetivos principais são:
- Maximizar a margem de separação entre as duas classes.
- Os pontos mais perto da margem de separação so os vetores de suporte. Estes pontos correspondem aos exemplos mais difíceis de classificar.
- Os vetores de suporte definem a margem de separação. Os pontos de treino fora da margem nao afetam a classificação.

<br>

As principais vantagens deste algoritmo são estimar o hiperplano ótimo de separação, lida bem com dados de alta dimensao e com problemas complexos. <br>
As principais desvantagens são ser necessario escolher apropriadamente a função de kernel e não lidar bem com quantidades elevadas de dados. <br><br>

Existem vários tipos de SVMs para classificação:
- LinearSVC: Classificador SVM com kernel linear
- SVC: Classificador SVM generico. Pode-se escolher vários tipos de funções de kernel.
- NuSVC: SVM generico modificado para a regularização ser controlada por um parâmetro ν (valores entre [0, 1]) em vez de C (valores entre [0,+∞[).
- SGDClassifier: SGD - Stochastic Gradient Descent - classificador (SVM ou discriminantelogístico). Convergência mais lenta, mas o método é util para lidar com grandes quantidades de dados.


O LinearSVC e o SVC são os dois que se destacam mais para o uso neste caso, e foram também os mais aprofundados em aula.<br>
Como tratamos de dados de texto, é expectavel que as características, ou seja, as palavras nos vários documentos, sejam linearmente separáveis. __Pelo que fica então a escolha entre usar o LinearSVC ou o SVC(kernel='linear').__ <br>

O LinearSVC é muito semelhante ao SVC com o parâmetro kernel='linear', mas implementado em termos de liblinear em vez de libsvm, pelo que tem mais flexibilidade na escolha de penalizações e funções de perda e é mais escalavél para um grande número de amostras.<br>
As principais diferenças entre LinearSVC e SVC residem na função de perda utilizada por padrão, e no tratamento da regularização de interceção entre estas duas implementações.
O suporte multiclasse é tratado de acordo com um esquema one-vs-rest enquanto o do SVC é tratado com one-vs-ones.<br>
É preciso também ter a noção que entre SVC e LinearSVC, __o LinearSVC tende a ser mais rápido a convergir__ para dados em larga escala, que é o nosso caso. <br><br>
Concluíndo, __o classificador usado será o LinearSVC__.

## KNN (K-Nearest Neighbors) vs Distancia ao centróide (classificadores por distância)

O KNeighborsClassifier é um classificador baseado em distância, ou seja, usa métricas de distância (falado na introdução teórica).
<br>
Em aula, foram dados dois tipos de classificadores baseados em distância:
- __Classificador de k vizinhos mais próximos__
- __Classificador de distancia ao centroide__

No classificador de distancia ao centroide, classifica-se uma nova observação (novo vector) baseado na distâncias às médias (centroides) das classes no conjunto de treino. A classe atribuída e a do centroide que estiver mais proximo do vector.

No classificador de k vizinhos mais próximos classifica-se uma nova observação baseado nas classes dos k vizinhos mais proximos do conjunto de treino. A classe atribuída por votação - classe maioritária nos k vizinhos.

Ambos os métodos têm seus pontos fortes e fracos, o tamanho do conjunto de dados e a da dimensionalidade dos dados e das características específicas dos documentos de texto em questão, influenciam o seu desempenho.<br>
No nosso caso, é expectável que o resultado sejam piores que os classificadores anteriores, tendo em conta que nas reviews de cervejas, muitas vezes não é tão claro a distinção quanto ao overall que se dá no final, por exemplo.<br>
De seguida, experimentar-se-á ambos os classificadores.



### Distância ao centróide - avaliação de muito boas e muito más

```python
def getBestParamsDistCentroide(ficheiro):

    D = pickle.load(open(ficheiro,'rb'))

    X_treino = D['X_treino']
    target_treino = np.array(D['target_treino'])
    X_teste = D['X_teste']
    target_teste = np.array(D['target_teste'])
    
    X_treino, X_valid, y_treino, y_valid = train_test_split(X_treino, target_treino, test_size=1/3, stratify=target_treino)
    
    #Calculo dos centroides/médias dos dados de treino 
    m0=np.squeeze(np.asarray(np.mean(X_treino[y_treino==0,:], axis=0)))
    m1=np.squeeze(np.asarray(np.mean(X_treino[y_treino==1,:], axis=0)))
    
    
    
    # Sendo que não é possível fazer diretamente a operação
    # D0=np.sqrt(np.sum((X2-m0[:,np.newaxis])**2,axis=0))
    # Porque não é possível passar da matriz para array devido ao espaço que ocupa
    
    # Sendo assim, em vez de fazemos a operação numa linha, percorremos documento a
    # documento, obtemos a distância para as diferentes classes e podemos logo 
    # obter a classe mais próxima e dar append no array das classes estimadas
  
    
    y2e = []
    
    for doc in X_valid:
        
        D0=np.sqrt(np.sum((doc[0].toarray()-m0)**2,axis=1))
        D1=np.sqrt(np.sum((doc[0].toarray()-m1)**2,axis=1))
        
        if(D0<D1):
            y2e.append(0)
        else:
            y2e.append(1)
    
    yte = []
    
    for doc in X_teste:
        D0=np.sqrt(np.sum((doc[0].toarray()-m0)**2,axis=1))
        D1=np.sqrt(np.sum((doc[0].toarray()-m1)**2,axis=1))
        
        
        
        if(D0<D1):
            yte.append(0)
        else:
            yte.append(1)
    
    #Representamos a matriz de cnfusão e o números total de erros nos dados de validação de teste
    matriz=confusion_matrix(y_valid,y2e)
    print('Matriz de confusão do grupo validação:\n',matriz)
    print('Taxa de erros no grupo de validação:',np.round(np.sum(y_valid!=y2e)/len(y2e)*100,2))
    print('Taxa de falsos positivos do grupo de validação:', np.round(matriz[0,1]/(matriz[0,1] + matriz[0,0])*100,2))
    print('Taxa de verdadeiros positivos do grupo de validação:', np.round(matriz[1,1]/(matriz[1,1] + matriz[1,0])*100,2))
    print('Score do grupo de validação:',np.round((matriz[1,1] + matriz[0,0])/len(y2e)*100,2))
    print()
    matriz=confusion_matrix(target_teste,yte)
    print('Matriz de confusão do grupo teste:\n',matriz)
    print('Taxa de erros no grupo de teste:',np.round(np.sum(target_teste!=yte)/len(yte)*100,2))
    print('Taxa de falsos positivos do grupo de teste:', np.round(matriz[0,1]/(matriz[0,1] + matriz[0,0])*100,2))
    print('Taxa de verdadeiros positivos do grupo de teste:', np.round(matriz[1,1]/(matriz[1,1] + matriz[1,0])*100,2))
    print('Score do grupo de teste:',np.round((matriz[1,1] + matriz[0,0])/len(y2e)*100,2))
    print()
    print()
```

O classificador k vizinhos mais proximos é um classificador supervisionado baseado em distâncias, onde não existe fase de treino. 

O __valor ótimo para o k__ dependem do problema, valores pequenos de k dão origem a zonas e fronteiras de decisão irregulares, valores muito elevados podem resultar em regiões e fronteiras demasiadas regulares.<br>
Quanto às weights, na "__uniform__", todos os pontos em cada vizinhança têm o mesmo peso, e na "__distance__", o peso dos pontos são o inverso da distância, ou seja, quanto mais perto maior o peso

```python
def getBestParamsKNeighborsClassifier(ficheiro, ficheiro_saida):
    
    D = pickle.load(open(ficheiro,'rb'))

    X_treino = D['X_treino']
    target_treino = D['target_treino']
    
    kArray = [5,10,15,20,25,30]
    wArray = ['uniform', 'distance']

    dicionarios={}
    count = 0
    for k in kArray:
        for w in wArray:
            
            print("Começou k:",k,'w:',w)
            
            knn = KNeighborsClassifier(n_neighbors=k, weights=w).fit(X_treino, target_treino)
            
            score = np.mean(cross_val_score(knn, X_treino, target_treino, cv=5))
                
            print("k: " + str(k) + ",W: " + str(w) + ",Score:" + str(score))
            resultados={'k' : k,
               'w' : w,
               'score': score}
            
            dicionarios[str(count)] = resultados
            count += 1
                
                
    pickle.dump(dicionarios, open(ficheiro_saida, 'wb'))
```

## Undersampling nas classificações binárias

Um ponto que foi várias vezes repetido na análise e conclusão das classificações binárias foi que com a calibração era possível aproximar o ponto operacional ao ótima com a consequência do aumento do número de erros.
__Mas porque é que é necessário fazer esta calibração?__ <br>
A verdade é que há uma questão que ainda não foi referida, que é o desíquilibrio entre dados classificados classificados como negativos e positivos tanto na classificação das muito más, como das muito boas.<br><br> __Classificando todas as respostas como negativas, obtem-se uma boa taxa de acertos, apesar de obviamente não ser o que se quer__ <br><br>
Podemos então optar por ter uma outra abordagem nos classificadores binários, que será fazer um undersampling e testar se os resultados obtidos são melhores ou piores.

Primeiramente, é necessário equilibrar os dados das muito boas e muito más.

```python
def getUndersampling(ficheiro, ficheiro_saida):

    D = pickle.load(open(ficheiro,'rb'))

    X_treino = D['X_treino']
    target_treino = np.array(D['target_treino'])
    X_teste = D['X_teste']
    target_teste = D['target_teste']
    
    tam_positivos = np.sum([target_treino == 1])
    
    array_positivos = X_treino[target_treino == 1]
    
    indexes = random.sample(range(75000), tam_positivos + 2000)
    
    X_treino = vstack((X_treino[indexes], array_positivos))
     
    target_treino = np.hstack((target_treino[indexes], target_treino[target_treino == 1]))
    
    print("Shape do novo grupo de treino:",X_treino.shape)
    print("Shape das classificações do novo grupo de treino:",target_treino.shape)
    
    dicionario = { 'X_treino': X_treino,
                 'target_treino':target_treino,
                 'X_teste': X_teste,
                 'target_teste':target_teste}
        
    pickle.dump(dicionario, open(ficheiro_saida, 'wb'))
```

```python
print("Grupo das muito más:")
getUndersampling('dados_treino_teste_muito_mas.p','dados_treino_reduzidos_teste_muito_mas.p')
print("Grupo das muito boas:")
getUndersampling('dados_treino_teste_muito_boas.p','dados_treino_reduzidos_teste_muito_boas.p')
```

```python
D = pickle.load(open('dados_treino_teste_multi.p','rb'))

X_treino = D['X_treino']
y_treino = D['target_treino']



num_class_8 = y_treino.count(8)

# indices of class 4, 5, 6, and 7 instances
indices_class_4 = [i for i in range(len(y_treino)) if y_treino[i] == 4]
indices_class_5 = [i for i in range(len(y_treino)) if y_treino[i] == 5]
indices_class_6 = [i for i in range(len(y_treino)) if y_treino[i] == 6]
indices_class_7 = [i for i in range(len(y_treino)) if y_treino[i] == 7]



# randomly sample instances from each of the classes 4, 5, 6, and 7 to match the number of instances for class 8
indices_to_remove = (
    indices_class_4[num_class_8:] +
    indices_class_5[num_class_8:] +
    indices_class_6[num_class_8:] +
    indices_class_7[num_class_8:]
)

# Remove the excess instances from classes 4, 5, 6, and 7
y1o_trimmed = [y_treino[i] for i in range(len(y_treino)) if i not in indices_to_remove]
reviews_train_trimmed = vstack(([X_treino[i,:] for i in range(len(y_treino)) if i not in indices_to_remove]))


D_multi = {}
D_multi['X_treino'] = reviews_train_trimmed
D_multi['X_teste'] = X_teste
D_multi['target_treino'] = y1o_trimmed
D_multi['target_teste'] = target_teste

pickle.dump(D_multi, open('dados_treino_teste_multi_trimmed.p', 'wb'))
```

## Regressão

A regressão é uma técnica para modelar e analisar dados que são compostos por uma variável dependente (a variável de resposta) e uma ou mais variáveis independentes (as variáveis de entrada). Ao contrário dos classificadores, o output dos métodos de regressão são valores contínuos. É esperado então que as formas de avaliar este método, não sejam iguais aos classificadores.
O objetivo da regressão é encontrar os melhores parâmetros para a equação que melhor descreve essa  relação. 
<br>
O objetivo principal é estimar os parâmetros (w1, w2 ...), de uma função paramétrica
$$ \hat y= w1φ1(x) + w2φ2(x) + . . . $$, com um conjunto com N amostras de entrada X = {x[1], . . . , x[N]}
e as correspondentes N amostras de saída Y = {y[1], . . . , y[N]}.
<br>
Existem dois tipos principais de regressão: __regressão linear e regressão não linear__.
<br><br>
A __regressão linear__  assume uma relação linear entre as variáveis, ou seja, a relação pode ser representada por uma linha reta. Para o caso de uma única variável independente.
A __regressão não linear__ é utilizada quando a relação entre as variáveis não pode ser adequadamente modelada por uma equação linear. Nesse caso, usa-se modelos matemáticos mais complexos, tal como as polinomiais.

### Regressão Linear

Quando falamos de regressão simples, a equação é da forma $$\hat y=mx+b$$, onde y é a variável dependente, x é a variável independente, m é a inclinação da linha e b é a interseção no eixo y.
<br>
Para os resultados obtidos, usar-se-ão 3 métodos:
- Coeficiente ${R^2}$ -  este coeficiente expressa a percentagem da variância da variável dependente que é predita pelas variáveis independentes.
- Erro quadrático médio - ${
\frac{1}{N}
\sum \limits _{n=1} ^{N}
(y[n] − (ax[n] +b))}$
- Erro absoluto médio - ${\frac{1}{N} \sum \limits _{n=1}^{N} |y[n] - \hat{y}[n]| }$

```python
lr = LinearRegression()

D = pickle.load(open('dados_treino_teste_muito_mas.p','rb'))

X_treino = D['X_treino']
X_teste = D['X_teste']
y_treino = D['target_treino']
y_teste = D['target_teste']

X_treino, X_valid, y_treino, y_valid = train_test_split(X_treino, y_treino, test_size=1/3, stratify=y_treino)
lr.fit(X_treino,y_treino)

W = lr.coef_.copy()
W[0] = lr.intercept_

ye_valid = lr.predict(X_valid)
ye_teste = lr.predict(X_teste)

print("R²:", lr.score(X_teste, y_teste))
print("Erro absoluto médio:", mean_absolute_error(y_teste, ye_teste))
print("Erro quadrático médio:", mean_squared_error(y_teste, ye_teste))
print('Valores dos pesos:\n', W)
```

### Regressão Polinomial

Para a regressão polinimial faz-se uso da função ___PolynomialFeatures___, que devolve a combinação polinomial das características, incluindo a coordenada homogénea.
<br><br>
Antes demais, é necessário ter em atenção que os termos polinomiais,que são o output da função anterior, serão:
$$
\frac{degree!×(numberofinputfeatures)!}
{(numberofinputfeatures+degree)!}
$$

Como é de esperar, se se utilizar o número de características que temos usado, o total, perto de 180 mil, o tamanho das matrizes vai aumentar muito. Depois de alguns testes feitos pelo grupo, percebemos que o prórpio jupyter apresenta um __Memory Error__ devido ao tamanho com que fica o output saído desta função.<br>
A solução passa então por fazer a vetorização novamente, com um número máximo de caracteristicas, e, caso necessário, com um número reduzido de dados também.
<br>
Por último, falta referir que a regressão polinomial pode facilmente entrar em sobre aprendizagem, particularmente em situações em que a dimensão dos dados é elevada  e o número de exemplos limitado (que é o nosso caso). Sendo assim, existem dois métodos que implementam a regressão linear com regularizaçãao dos pesos: __Ridge__ e __Lasso__.


```python
Dic = pickle.load(open('Docs_with_stemming.p','rb'))
Docs_train    = Dic['Docs_train']
Docs_test     = Dic['Docs_test']
target_treino = Dic['target_treino']
target_teste  = Dic['target_teste']

target_treino_negativo = [0 if num >= 3 else 1 for num in target_treino]    
target_teste_negativo = [0 if num >= 3 else 1 for num in target_teste]

target_treino_positivo = [0 if num <= 8 else 1 for num in target_treino]    
target_teste_positivo = [0 if num <= 8 else 1 for num in target_teste]

tfidf=TfidfVectorizer(min_df=3, token_pattern=r'\b\w\w+\b', ngram_range=(1,2), max_features=250).fit(Docs_train) 
X_treino = tfidf.transform(Docs_train)[:4000]
X_teste = tfidf.transform(Docs_test)[:4000]

print("Shape da vetorização dos documentos de treino:",X_treino.shape)
print("Shape da vetorização dos documentos de teste:",X_teste.shape)

D_muito_mas = {}
D_muito_mas['X_treino'] = X_treino
D_muito_mas['X_teste'] = X_teste
D_muito_mas['target_treino'] = target_treino_negativo[:4000]
D_muito_mas['target_teste'] = target_teste_negativo[:4000]


D_muito_boas = {}
D_muito_boas['X_treino'] = X_treino
D_muito_boas['X_teste'] = X_teste
D_muito_boas['target_treino'] = target_treino_positivo[:4000]
D_muito_boas['target_teste'] = target_teste_positivo[:4000]

```

```python
X_treino = D_muito_boas['X_treino']
X_teste = D_muito_boas['X_teste']
y_treino = D_muito_boas['target_treino']
y_teste = D_muito_boas['target_teste']

print(X_treino.shape)

print("Começou")
poly = PolynomialFeatures(degree=4, interaction_only=True).fit(X_treino)

X1b  = poly.transform(X_treino)
print("Fez 1")
X2b  = poly.transform(X_teste)
print("Fez 2")

#Utilizando a função disponibilizada pelo sklearn
lr = Ridge(alpha=1.0).fit(X1b,y_treino)
W = lr.coef_.copy()
W[0] = lr.intercept_

print('Valores dos pesos:\n', W)

# Utilizando também o sklearn para calcular o coeficiente de determinação
print('Treino:', lr.score(X1b, y_treino))
print('Teste:' , lr.score(X2b, y_teste))
```

## PCA

Apesar dos grupo de documentos serem de 75 mil, no caso do grupo de treino, e 25 mil no grupo de teste, há um aumento exorbitante desta dimensionalidade quando fazemos a vetorização, onde os dados passam a ser matrizes (nº de documentos x nº de tokens). <br>
Nas classificações feitas anteriormente, foram usados todas as características disponíveis, mas, numa outra abordagem, podia-se tentar diminui-las, o que ajudaria a ter um menor processamento na altura de treinar o classificador, e até poderia ajudar a reduzir a sobreaprendizagem.<br><br>
Aqui entra o __PCA__, que é justamente um mecanismo de redução da dimensionalidade de dados de elevada dimensão. O PCA mais especificamente reduz a dimensionalidade de maneira a preservar o mais possível a variação presente nos dados de alta dimensão, ou seja, projectar dados nas direcções de maior variância, sendo que as direcções onde os dados variam mais contem mais informação. Inicialmente, o número de caracteristicas é igual ao número de características originais, no entanto, o objetivo é selecionar um subconjunto menor de componentes principais que capture a maior parte da variância, eliminando as com menor contribuição para a variância total. O PCA trata-se também de um método não supervisionado.
<br> 
Para além do PCA, aprendeu-se também em aula outro mecnismo de redução da dimensionalidade, o __LDA__, que reduz a dimensionalidade de maneira a preservar o mais possível a informação discriminativa entre classes presentes nos dados de alta dimensão.

### Classificações Binárias

```python
def getBestCompsPCA(ficheiro, ficheiro_saida):
    
    D = pickle.load(open(ficheiro,'rb'))

    X_treino = D['X_treino']
    target_treino_muito_mas = D['target_treino']
    
    print(X_treino.shape)
    
    CompsArray = [100, 200, 300, 400, 500]
        
    dicionarios={}
    count = 0
    for C in CompsArray:
        
        svd = TruncatedSVD(n_components=C, n_iter=5)
        X_treino_SVD = svd.fit_transform(X_treino)
        
        lr = LogisticRegression(penalty='l2', solver='liblinear', C=2.5)
        lr.fit(X_treino_SVD, target_treino_muito_mas)
    
        score = np.mean(cross_val_score(lr, X_treino_SVD, target_treino_muito_mas, cv=5))
                
        print("Nº Componentes: " + str(C) + ",Score:" + str(score))
        resultados={'Comp' : C,
            'score': score}
            
        dicionarios[str(count)] = resultados
        count += 1
                
                
    pickle.dump(dicionarios, open(ficheiro_saida, 'wb'))
```

## Clustering

Os metodos de agrupamento, ou clustering, são técnicas de aprendizagem não supervisionada onde se procura encontrar representaçõoes mais compactas dos dados sem mais nenhuma informação sobre os mesmos. O objetivo é dividir os dados (vetores d-dimensionais) em grupos, de modo a que elementos do mesmo grupo sejam mais semelhantes entre si do que com membros de outros grupos. Isto implica definir uma função de semelhança e maximizar este valor mudando a atribuição dos grupos a cada ponto.
<br>
O cenario ideal é os dados estarem agrupados em nuvens multi-dimensionais de pontos, e nestas condições é provavel que as observações pertencentes a um dado cluster tenham características comuns entre elas. Desta forma pode-se categorizar os dados e
descobrir as “classes” que neles estao subjacentes, obtendo assim uma representação de alto nível dos dados. 
<br>
Existem vários tipos de algoritmos de clustering, mas teremos mais atenção a dois:
- K médias
- Dendrogramas



### K-médias

O algoritmo k-medias é um dos mais simples em aprendizagem automática. Esta técnica representa os dados através de k centroides em que cada ponto pertence ao centroide mais proximo. Para isto, utiliza-se a função ___KMeans___. 

Primeiramente, é necessário ter em atenção a escolha do __n_clusters__, ou seja, do número de centróides.<br>
Existem dois métodos comuns para descobrir o número ótimo:
- __Método do Cotovelo__ (Elbow Method)
- __Método da Silhueta__ (Silhouette Method)

No método do cotovelo, procura-se o ponto onde a variância começa a nivelar, formando um "cotovelo". Este ponto é geralmente considerado um bom valor para o número de clusters.
<br>
No método da silhueta, o valor de k que maximiza a métrica de silhueta é considerado o número ideal de clusters.<br>
Pode-se então testar com ambos os métodos e decidir a partir da sua analise.

### Dendrogramas

Dendrogramas são árvores de relação hierárquica de proximidade entre pontos. A estratégia é juntar os dois pontos mais próximos, e repetir o processo entre pontos ou conjuntos de pontos já juntos até não haver mais pontos para juntar.<br>
O algoritmo utilizado para a formação dos dendogramas é:
- considera-se cada ponto um cluster
- encontra-se os dois clusters mais semelhantes
- junta-se os dois clusters num único
- calcula-se a distância deste novo cluster com os restantes
- repete-se até só haver um cluster

Existem vários métodos de ligação:
- Ligação simples - menor distância entre dois pontos
- Ligação completa - maior distância entre dois pontos dos dois conjuntos
- Ligação média - média de distâncias entre todos os pontos dos dois conjuntos
- Ligação ao centroid - distância entre as médias dos dois conjuntos

Para implementar os dendogramas, usa-se o objeto __AgglomerativeClustering__.<br>
Existem duas abordagens possíveis de ter nesta implementação:
- pelo n_clusters
- pelo distance_threshold, o algoritmo forma clusters até que a distância entre os pontos seja maior que o valor especificado.

```python
Dic = pickle.load(open('Docs_with_stemming.p','rb'))
Docs_train    = Dic['Docs_train']

tfidf=TfidfVectorizer(min_df=3, token_pattern=r'\b\w\w+\b', ngram_range=(1,2), max_features=100, max_df=0.95, stop_words='english').fit(Docs_train[:1000]) 
X_treino = tfidf.transform(Docs_train[:1000])

distance_thresholds = []
silhouette_scores = []

for threshold in np.linspace(0.1, 4.0, 40):
    print("Distância:",threshold)
    clustering = AgglomerativeClustering(n_clusters=None, distance_threshold=threshold).fit(X_treino.toarray())
    labels = clustering.labels_
    unique_labels = len(set(labels))
    if unique_labels > 1 and unique_labels < len(X_treino.toarray()):
        distance_thresholds.append(threshold)
        silhouette_scores.append(silhouette_score(X_treino.toarray(), labels))
    

```

```python
# Plotagem do método do cotovelo
plt.plot(distance_thresholds, silhouette_scores, marker='o')
plt.xlabel('Distance Threshold')
plt.ylabel('Inertia')
plt.title('Método do Cotovelo para Clustering Hierárquico Aglomerativo')
plt.show()
```

Para os dendogramas, avalia-se apenas pelo método da silhueta. Novamente, os valores não parecem ser os melhores, como foi explicado anteriormente no kmédias. <br>
Mesmo assim, observando o gráfico é claro um valor que maximiza este resultado __distance_threshold = 1.2__.

```python
distance_threshold = 1.2

terms = tfidf.get_feature_names()

# Clustering Hierárquico Aglomerativo
clustering = AgglomerativeClustering(n_clusters=None, distance_threshold=distance_threshold).fit(X_treino.toarray())
labels = clustering.labels_

# Clustering Hierárquico Aglomerativo
linked = linkage(X_treino.toarray(), 'ward')
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrograma de Clustering Hierárquico')
plt.show()
```

Tendo disponível o dendrograma dos pontos, pode-se visualmente aferir onde se deve "cortar" a árvore para obter os "melhores" clusters.<br>
